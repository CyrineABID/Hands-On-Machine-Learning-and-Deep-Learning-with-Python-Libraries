# Libraries

In Python, a library is a **collection** of pre‑written code (modules and packages) that you can import into your program to avoid reinventing the wheel. Instead of writing everything from scratch, you use **libraries** to access ready‑made functions, classes, and tools.

A **module** is a single file of Python code (e.g., math.py). [math](https://docs.python.org/3/library/math.html) <br>
A **package** is a collection of modules organized together (e.g., numpy, pandas). [pandas](https://pandas.pydata.org/) <br>
A **library** is a broader term that often refers to a set of packages/modules designed for a specific purpose (e.g., Scikit‑learn for machine learning, TensorFlow for deep learning). [Scikit-learn](http://scikit-learn.org/stable/) | [TensorFlow](https://www.tensorflow.org/?hl=fr)

The essential Python libraries for **Machine Learning** (ML) and **Deep Learning** (DL) include NumPy, Pandas, SciPy, Scikit‑learn, TensorFlow, and PyTorch. <br>
For **specialized** tasks, you’ll also use **libraries** like Keras, XGBoost, LightGBM, OpenCV, NLTK, and SpaCy.

# 1. Dataframe manipulation

## 1.1. Numpy

## What is NumPy?
**NumPy (Numerical Python)** is a core Python library for numerical computing.  
It provides:
- **ndarray** → A powerful n-dimensional array object for storing and manipulating numerical data.
- Tools for mathematical operations, linear algebra, statistics, and more.
- The foundation on which **Pandas** is built — Pandas DataFrames use NumPy arrays internally.

---

## Getting Started
Install NumPy (if not already installed):
```bash
pip install numpy

````python
import numpy as np
````

**1. Creating Arrays**

````python
# 1D array
arr = np.array([1, 2, 3, 4, 5])

# 2D array (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
````

**2. Array Operations**

````python
arr + 10        # element-wise addition
arr * 2         # element-wise multiplication
print("the average is", np.mean(arr))   # average
print("the std is ", np.std(arr))     # standard deviation
````

**3. Indexing and Slicing**

````python
arr[0]      # first element
arr[1:4]    # slice from index 1 to 3
matrix[0,2] # element at row 0, column 2
````

## Why NumPy?
* *Speed*: Much faster than Python lists for numerical operations.
* *Foundation*: Pandas DataFrames are built on NumPy arrays.
* *Essential for ML/DL*: Libraries like Scikit-learn, TensorFlow, and PyTorch rely on NumPy.

## 1.2. Pandas

## What is Pandas?
**Pandas** is a powerful Python library designed for **data analysis and manipulation**.  
It provides two main data structures:
- **Series** → A one-dimensional labeled array (like a column in Excel).
- **DataFrame** → A two-dimensional labeled data structure (like a table in Excel).

Pandas makes it easy to:
- Load data from files (CSV, Excel, SQL, JSON, etc.)
- Clean and preprocess datasets
- Select, filter, and transform data
- Perform statistical analysis
- Visualize data with integration to libraries like Matplotlib and Seaborn

---

## Getting Started
First, install Pandas (if not already installed):
```bash
pip install pandas


````python
import pandas as pd #import it in the Jupyter notebook
````

## Basic Usage
**1. Creating a DataFrame**

````python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["Paris", "London", "Berlin"]
}

df = pd.DataFrame(data)
df
````

**2. Selecting Columns and Rows**

````python
df["Name"]        # Select a column
df.iloc[0]        # Select first row
df.loc[1, "City"] # Select specific cell
````

**3. Basic Operations**

````python
df.describe()     # Summary statistics
df.sort_values("Age")  # Sort by Age
df[df["Age"] > 28]     # Filter rows
````

## Why Pandas?
* Intuitive and beginner-friendly
* Handles real-world messy datasets
* Essential for machine learning workflows (data preprocessing before modeling)

## What is TensorFlow?

---

## 🔹 The Sequential Model
Before starting deep learning, it’s important to understand the **Sequential model** in Keras:

- A **Sequential model** is a **linear stack of layers**.
- Data flows forward from the input layer → hidden layers → output layer.
- It’s the easiest way to build a neural network when your architecture is straightforward.

### Example:
```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define a simple feedforward neural network
model = Sequential([
    Dense(32, activation='relu', input_shape=(784,)),  # Input + first hidden layer
    Dense(16, activation='relu'),                      # Second hidden layer
    Dense(10, activation='softmax')                    # Output layer (10 classes)
])


````python
from tensorflow import keras 
````

### What is the Iris Dataset?
- The **Iris dataset** is one of the most famous datasets in machine learning.
- It contains **150 samples** of iris flowers, divided into **3 species**:
  - *Iris setosa*
  - *Iris versicolor*
  - *Iris virginica*
- Each sample has **4 features** (measurements):

  - Sepal width
  - Petal length
  - Petal width

---

### Why Iris?
- Small and simple → easy to visualize and understand.
- Perfect for learning **classification algorithms**.
- Often used as a beginner dataset to test models like logistic regression, decision trees, k-nearest neighbors, and support vector machines.

---

# Introduction to cBioPortal and the BRCA_MSK_2025 Dataset

## What is cBioPortal?
- **cBioPortal** is a free, open‑source platform for exploring **cancer genomics data**.
- It integrates large‑scale datasets from projects like TCGA (The Cancer Genome Atlas) and MSK (Memorial Sloan Kettering Cancer Center).
- Provides interactive tools to:
  - Visualize genetic alterations (mutations, copy number changes, fusions).
  - Explore clinical data alongside molecular profiles.
  - Perform queries across thousands of cancer samples.

Think of cBioPortal as a **search engine and visualization tool for cancer data**. <br>
In fact, researchers and students use it to:
* Visualize mutations, copy number changes, and other genomic alterations.
* Explore patient clinical data (diagnosis, tumor type, outcomes).
* Connect genomic patterns with clinical features.

---

## What is the BRCA_MSK_2025 Dataset?
- **BRCA_MSK_2025** is a breast cancer study published in *NPJ Precision Oncology (2025)* by Memorial Sloan Kettering (MSK).
- Int includes **527 breast carcinoma samples** from **462 patients**.
- The focus is on **tumor mutational burden (TMB-high breast carcinomas)** - tumors with a high number of mutations.
- It contains:
  - **Genomic characterization**: mutation counts, fraction of genome altered, copy number segments.
  - **Clinical data**: cancer type (ductal, lobular, mixed), whether the sample is primary or metastatic, and patient/sample IDs.
- The dataset is designed to help researchers and students:
  - Understand the genetic landscape of breast cancer.
  - Explore how genomic changes relate to clinical outcomes.
  - Practice bioinformatics and machine learning methods on real cancer data.

---

## Why is it Important?
- **Educational value**: Beginners can learn how to connect clinical features (like tumor type) with genomic data (mutation counts).
- **Research relevance**: Helps identify how mutation burden relates to breast cancer progression and treatment response.
- **Hands‑on practice**: Great dataset for applying machine learning methods (classification, clustering, survival analysis).

---

## Connection to Machine Learning <br>
This dataset can be used to :
* Classification : Predict tumor subtype (ductal vs lobular vs mixed)
* Regression : Model mutation count or fraction genome altered or TMB
* Clustering : Group patients by genomic profiles

## **3. For Deep Learning Example**

***The MNIST Digit Dataset***

### What is MNIST?
- **MNIST** stands for *Modified National Institute of Standards and Technology* dataset.
- It contains **70,000 grayscale images** of handwritten digits (0–9).
- Each image is **28x28 pixels**.
- Widely used as a beginner dataset for testing and learning deep learning models.

---

###  Why MNIST?
- Simple and small → easy to train models quickly.
- Standard benchmark → many tutorials and papers use it.
- Perfect for learning how to build and evaluate neural networks.

---

###  Loading MNIST in TensorFlow/Keras
Keras provides MNIST directly, so you don’t need to download it manually:

```python
from tensorflow.keras.datasets import mnist

# Load dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("Training data shape:", x_train.shape)
print("Test data shape:", x_test.shape)
```

---

### Preprocessing <br>
Before feeding into a neural network:
* Normalize pixel values (0–255 → 0–1).
* Flatten images (28x28 → 784 features) if using Dense layers.
* One-hot encode labels for classification.

---

### Connection to Sequential Model <br>
The MNIST dataset is often the first dataset used with a Sequential model:
* Input layer → 784 features (flattened pixels).
* Hidden layers → Dense layers with ReLU activation.
* Output layer → 10 neurons (digits 0–9) with softmax activation.

This makes MNIST the perfect playground to understand how data flows through a neural network.
