## How to handle loops in Python ?

```python
# Looping through a range of numbers
for i in range(5):  # i is loop counter or iterator
    print("Iteration:", i) #display interation value

# Looping through a list
fruits = ["apple", "banana", "cherry"] #define a list : collection of items
for fruit in fruits:
    print("I like", fruit)  #display the list items
```

## Conditional Statements

```python
number = 10      #starting point
if number > 0:   #Condition should be correct
    print("Positive number")
elif number == 0: #if not display second value
    print("Zero")
else:  #if not  display third value
    print("Negative number")
```

## How to define a new funtion ?

```python
def greet(name):   #greet is the name of the function and name is a necessary argument to define by the user
    return f"Hello, {name}!"  #value to return after executing the function

print(greet("GTCA Network"))
```

## Core Python Data Structures

## 1.List
Ordered, mutable collection created by placing items inside square brackets [ ], separated by commas

```python
fruits = ["apple", "banana", "cherry"]
```

```python
print (fruits)
```

```python
print(fruits[2])
```

## 2. Dictionary
Key–value pairs, like a mini database using curly braces { }

```python
person = {"Project": "GTCA", "Training": "BEST", "Edition": 2}
```

```python
print ("Our", person['Project'], "project", "has a training program called", \
       person["Training"], "and its already in its", person["Edition"], "edition")
```

## 3. Tuple
Ordered, immutable collection (cannot be changed after creation). To define a tuple, parentheses ( ) are used

```python
coordinates = (10, 20) #start indexing from 0
print (coordinates)
```

```python
print(coordinates[1])
```

## 4. Set
Unordered collection of unique items (no duplicates).

```python
colors = {"red", "green", "blue"}
print (colors)
```

```python
print ("red" in colors) #membership test
```
